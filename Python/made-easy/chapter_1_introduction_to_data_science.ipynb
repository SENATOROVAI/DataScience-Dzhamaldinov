Чем занимается специалист по обработке данных?
Специалист по обработке данных анализирует бизнес-данные, чтобы извлечь значимую информацию. Другими словами, специалист по обработке данных решает бизнес-задачи с помощью ряда шагов:

- Задает правильные вопросы, чтобы понять проблему
- Собирает данные из нескольких источников — корпоративных данных, общедоступных данных и т.д
- Обрабатывает необработанные данные и преобразует их в формате, подходящем для анализа
- Вводит данные в алгоритм analytical system — ML или статистическую модель
- Подготовливает результаты и аналитическую информацию для обмена с соответствующими заинтересованными сторонами

Это помогает решить множество важных вопросов, некоторые из которых приведены ниже:
- Выявление мошенничества и аномалий, например изменение порядка снятия средств с кредитной карты клиента или их расходования.
- Целевой и персонализированный маркетинг — рассылка персональных сообщений по электронной почте, рекомендательные системы на веб-сайтах электронной коммерции.
- Прогнозы показателей — показателей эффективности, качества рекламных кампаний и других мероприятий.
- Скоринг принятия решений — обрабатывает большие объемы данных и помогает принять решение, например, о предоставлении кредита, на основе кредитных баллов.
- Прогнозирующее моделирование - прогнозирование столкновения метеорита с землей на основе астрономических данных.

Компилятор
Компилятор преобразует исходный код в машинный.
Компилятор переводит или, скорее, компилирует исходный код с языка программирования высокого уровня на машинный язык.

Интерпретатор
В компьютерном программировании интерпретатор - это компьютерная программа, которая непосредственно выполняет инструкции, написанные на языке программирования, не требуя, чтобы они были предварительно скомпилированы в программу на машинном языке.
По сути, это то, что делают компьютерные интерпретаторы: они обрабатывают ваш исходный код при каждом запуске, строка за строкой, и пользователю необходимо, чтобы на его компьютере был доступен необходимый интерпретатор.

И компиляторы, и интерпретаторы преобразуют исходный код (текстовые файлы) в машинный язык. Основное различие заключается в том, что система компилятора генерирует автономную программу с машинным кодом, в то время как система интерпретатора выполняет действия, описанные программой высокого уровня.

После компиляции программы для ее запуска не требуется ее исходный код.
Для интерпретируемых программ исходный код необходим для каждого запуска программы.

Как правило, интерпретируемые программы выполняются медленнее, чем скомпилированные программы.



1.  Какие предметные области входят в Data Science? Что между ними общего и 
в чем различие?



Data Science - это междисциплинарная область, которая объединяет множество дисциплин. Вот некоторые из наиболее распространенных:

1. Статистика:
Общее:  Анализ и интерпретация данных, использование вероятностных моделей для прогнозирования, поиск взаимосвязей и закономерностей.
Различие:  Фокус на выводе обобщений из выборок, проверке гипотез и оценке неопределенности.

2. Машинное обучение (Machine Learning):
Общее:  Обучение алгоритмов на данных для автоматизации принятия решений, прогнозирования и классификации.
Различие:  Применение алгоритмов, которые учатся на данных без явного программирования. 

3. Глубокое обучение (Deep Learning):
Общее:  Подмножество машинного обучения, использующее искусственные нейронные сети с множеством слоев для обработки сложных данных.
Различие:  Способность справляться с неструктурированными данными, такими как изображения и текст.

4. Информатика:
Общее:  Создание и управление алгоритмами, структур данных, баз данных и вычислительных систем.
Различие:  Фокус на эффективных методах хранения, обработки и доступа к большим объемам данных.

5. Визуализация данных:
Общее:  Представление данных в визуальной форме для лучшего понимания и коммуникации.
Различие:  Создание интерактивных графиков, диаграмм и карт для исследовательского анализа и представления результатов.

6. Доменная экспертиза:
Общее:  Понимание специфики области, для которой используются данные.
Различие:  Позволяет создавать более релевантные решения и интерпретировать результаты в контексте проблемы.

## Общее между предметными областями:

Использование данных:  Все области Data Science работают с данными.
Поиск закономерностей:  Все области стремятся найти закономерности и информацию в данных.
Решение проблем:  Все области используются для решения реальных проблем с помощью данных.

## Различия между предметными областями:

Фокус: Каждая область имеет свой уникальный фокус и используемые методы.
Уровень абстракции:  Некоторые области работают с более абстрактными понятиями, чем другие.
Инструменты и технологии:  Каждая область использует свой набор инструментов и технологий.


Data Science - это синтез этих областей, который позволяет решать сложные проблемы с помощью данных.  Умение объединять знания из разных областей является ключевым для успешной работы в Data Science.



2.  Как вы понимаете термин «алгоритм»? Как алгоритмы связаны с блок-схемами?

Алгоритм - описанная на некотором языке точная конечная система правил, определяющая содержание и порядок действий над некоторыми объектами, строгое выполнение которых дает решение поставленной задачи.Понятие алгоритма, являющееся фундаментальным в математике и информатике.



Блок-схема алгоритма — графическое изображение алгоритма в виде связанных между собой с помощью стрелок (линий перехода) и блоков — графических символов, каждый из которых соответствует одному шагу алгоритма. Внутри блока дается описание соответствующего действия.

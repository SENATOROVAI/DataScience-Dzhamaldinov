Python - это свободный, интерпретируемый язык программирования высокого уровня общего назначения с открытым исходным кодом.

<h4>CPython</h4>

Это стандартная и наиболее широко используемая реализация языка программирования Python.
Напи­санный на С и Python, интерпретатор CPython содер­жит много внешних функций С и связан с другими 
языками программирования.
CPython можно в какой-то степени считать компиля­тором,поскольку перед интерпретацией
он преобразует код Python в байт-код. В нем используется GIL (глобальная блокировка интерпретатора),
которая ограни­чивает возможности интерпретатора, поскольку она отключает параллельные
потоки Python в рамках одного процесса. Будучи эталонной реализацией Python,
CPython поддерживает максимальную со­вместимость с пакетами Python и модулями
расширения С. Таким образом, все вер­сии языка программирования Python реализованы на С.
CPython— единственный вариант использования пакетов Python, которому для 
правильной работы нужны расширения С. Использование CPython совершенно необходимо,
если ваша программа на Python должна охватить максимально широкую аудиторию.






Pycach нужен чтобы сохранить уже проверенный Мини-компилятором код для повторного использования.

<h4>Jython</h4>

Ранее известная как JPython, реализация Jython работает на платформе Java.
Это сочетание Java и Python преобразует код Python в байт-код Java и, следовательно,
позволяет за­пускать код Python на любой машине, на которой установ­лена JVM.

Jython обеспечивает поддержку как статической, так и динамической компиляции.
Важной особенностью этого интерпретатора Python является то, что он позволяет
импор­тировать, а также использовать любые классы Java так, словно это модуль Python.
Если вам требуется взаимодействие с существующим кодом на Java или нужно
на­писать код Python для JVM, Jython — ваш кандидат.


<h4>РуРу</h4>

РуРу — это быстрая и совместимая альтернативная реализация языка программирования Python.

Благодаря JIT-компилятору РуРу поддерживает серверные части С, CLI и JVM.
Основная цель РуРу — обеспе­чить максимальную совместимость с эталонной реализацией CPython и в то же 
время увеличить производительность. РуРу используется разработчиками Python,
которые хотят увеличить производи­ тельность некоторых видов кода Python.

PythonNet можно использо­вать совместно с IronPython без каких-либо проблем. 


<h4>Stackless Python</h4>

CPython и другие популярные интерпретаторы Python для управления стеком используют язык С.
Но к интерпретатору Stackless Python это не относится.
Хотя в Stackless Python используется стек С, между вызовами функций он очищается.
Следовательно, интерпретатор Python в работе со стеком не зависит от вызова С. Как и CPython, 
Stackless Python написан с использованием С и Python.

Помимо поддержки потоков, у Stackless Python есть поддержка 
каналов связи, сопрограмм, предварительно скомпилированных двоичных файлов, 
циклического планирования, сериализации задач и тасклетов.
Возможно, самая важная особенность Stackless Python — это наличие микропото­ков.
Эта возможность помогает избежать значительной части накладных расходов, 
связанных с обычными потоками операционной системы.

<h4>Ячейки Markdown</h4>

Вы можете грамотно документировать вычислительный процесс, чередуя описательный текст с кодом, используя форматированный текст. В IPython это достигается путем разметки текста с помощью языка Markdown. Соответствующие ячейки называются ячейками Markdown. Язык Markdown предоставляет простой способ выполнить такую разметку текста, то есть указать, какие части текста следует выделять курсивом, жирным шрифтом, формировать списки и т.д.

Если вы хотите структурировать свой документ, вы можете использовать заголовки markdown. Заголовки Markdown состоят из 1-6 хэш-символов #, за которыми следует пробел и название вашего раздела. Заголовок markdown будет преобразован в интерактивную ссылку на раздел записной книжки. Он также используется в качестве подсказки при экспорте в другие форматы документов, такие как PDF.

При выполнении ячейки Markdown код Markdown преобразуется в соответствующий форматированный форматированный текст. Markdown допускает форматирование произвольного HTML-кода.



<h4>1. Python — это программное обеспечение с открытым исходным кодом. Это то 
же самое, что и бесплатное ПО?</h4>

- Python легко модифицируется сообществом разработчиков
Прежде всего, Python является бесплатным для использования. Это означает, что вы можете скачивать Python, писать код на Python и даже бесплатно продавать код на Python кому угодно. Это ключевой момент в использовании открытого исходного кода.

Кроме того, у вас есть прямой доступ как к просмотру, так и к изменению кода Python. Это проект CPython core на GitHub. Если вы заметили проблему с языком или захотели его изменить, вы можете отправить запрос на обновление и внести свой вклад в разработку языка! Однако участие в проекте, очевидно, потребует тщательной проверки, и это будет очень сложно сделать. Несмотря на это, такая возможность есть. Вы можете вносить дополнения в их документацию, исправлять ошибки, добавлять модульные тесты или делать практически все, что улучшит язык Python.

Именно это делает открытый исходный код таким мощным. Он опирается на сообщество разработчиков, тех, кто
непосредственно использует его, чтобы сделать его отличным!

Язык, который закрыт для использования кем бы то ни было, на самом деле является закрытым языком.

Закрытые языки обычно существуют в частных компаниях, которые не открыты для публичного использования.

- Python разработан по лицензии, одобренной OSi 
Если вам нужны более конкретные доводы в пользу того, почему Python имеет открытый исходный код, мы можем обратить внимание на тот факт, что он разработан по лицензии, одобренной OSi.

Это означает, что по закону всем разрешено неограниченное использование, модификация и распространение Python. Лицензия должна пройти процедуру проверки лицензий Open Source Initiative, чтобы быть принятой OSI.

Чтобы язык был сертифицирован OSI, он должен быть доступен по лицензии, гарантирующей право на использование, чтение и модификацию программного обеспечения. Python в свою очередь прошел тест OSI. 

- Авторские права на Python защищены лицензией, совместимой с GPL

<h4>7. В чем разница между «расширяемым» и «встраиваемым» языком?</h4>

1. Расширяемый язык (Extensible language):
   - Позволяет пользователю добавлять собственные функции, операторы и синтаксические конструкции.
   - Пользователь может расширять язык, чтобы лучше соответствовать своим потребностям и задачам.
   - Примеры расширяемых языков: Lisp, Tcl, Emacs Lisp.

2. Встраиваемый язык (Embedded language):
   - Это язык, который встраивается или интегрируется в другое программное обеспечение или приложение.
   - Встраиваемый язык обычно используется для расширения функциональности и настройки основного приложения.
   - Примеры встраиваемых языков: Lua, Python, Perl (в качестве встраиваемых скриптовых языков в различных приложениях).

Основная разница заключается в том, что расширяемые языки позволяют пользователю расширять сам язык, в то время как встраиваемые языки встраиваются в другие приложения и используются для расширения и настройки этих приложений.

Расширяемые языки обычно предоставляют более гибкую и мощную среду для разработки, позволяя программистам определять собственные конструкции языка. Встраиваемые языки, с другой стороны, предназначены для интеграции с другими системами и приложениями, обеспечивая возможность расширения и настройки без необходимости изменения основного кода.


<h4>8. В чем смысл IDE? Чем она отличается от командной строки?</h4>

1. Графический интерфейс:
   - IDE предоставляет графический интерфейс пользователя с различными панелями, окнами и инструментами.
   - Командная строка имеет текстовый интерфейс без графических элементов.

2. Интегрированные инструменты:
   - IDE объединяет в себе множество инструментов для разработки, включая редактор кода, компилятор, отладчик, менеджер проектов и т.д.
   - Командная строка требует использования отдельных командных утилит для выполнения различных задач.

3. Удобство работы:
   - IDE обеспечивает более удобную и интуитивную среду для разработки, с автодополнением кода, подсветкой синтаксиса, визуальными инструментами и т.п.
   - Командная строка требует большего ручного ввода команд и может быть менее интуитивной для начинающих разработчиков.

4. Управление проектами:
   - IDE часто включает средства для управления проектами, такие как системы контроля версий, сборщики, развертывание и т.д.
   - Командная строка требует использования отдельных инструментов для управления проектами.

5. Производительность:
   - IDE предоставляет комплексную среду, что может повысить производительность опытных разработчиков.
   - Командная строка может быть более гибкой и подходить для конкретных задач или опытных пользователей.

В целом, IDE помогает повысить эффективность и удобство разработки, особенно для начинающих разработчиков, в то время как командная строка предоставляет больше гибкости и контроля для более опытных пользователей.


<h4>9. Как открыть существующий документ Jupyter Notebook? Чем эта процедура от­
личается от открытия PDF-файла или текстового файла?</h4>

1. Запустите Jupyter Notebook на вашем компьютере. Это можно сделать, запустив Jupyter Notebook из командной строки или из приложения, которое вы используете для запуска Jupyter Notebook.

2. Когда Jupyter Notebook запустится, в браузере откроется интерфейс Jupyter Notebook, отображающий ваш рабочий каталог.

3. Найдите существующий файл Jupyter Notebook (обычно имеющий расширение `.ipynb`) в вашем рабочем каталоге или в другом каталоге, к которому вы имеете доступ.

4. Щелкните по файлу Jupyter Notebook, чтобы открыть его.

Это отличается от открытия PDF-файла или текстового файла следующим образом:

1. PDF-файлы и текстовые файлы обычно открываются с помощью специализированных приложений, таких как Adobe Reader для PDF-файлов или текстовые редакторы для текстовых файлов.

2. Файлы Jupyter Notebook открываются непосредственно в браузере, так как Jupyter Notebook является веб-приложением, работающим на вашем локальном компьютере.

3. Файлы Jupyter Notebook содержат интерактивный код и могут выполняться ячейка за ячейкой, в то время как PDF-файлы и текстовые файлы являются статическими документами.


<h4>10. В чем разница между «ячейками разметки Markdown» и «ячейками кода» 
в Jupyter Notebook? Для чего они нужны?</h4>

Ячейки разметки Markdown:
- Эти ячейки используются для текста, заголовков, списков, формул, изображений и другого форматированного контента.
- Они позволяют включать в Jupyter Notebook текстовую документацию, пояснения и описания в формате Markdown.
- Markdown - это облегченный язык разметки, который используется для форматирования текста в простой и читаемой форме.
- Ячейки Markdown отображают форматированный текст при выполнении ячейки.
- Они служат для создания читаемых и структурированных документов внутри Jupyter Notebook.

Ячейки кода:
- Эти ячейки предназначены для написания и выполнения программного кода.
- Они позволяют вводить, редактировать и выполнять код на языках программирования, таких как Python, R, Julia и другие.
- При выполнении ячейки кода, код в ней будет обработан, а результаты будут отображены ниже ячейки.
- Ячейки кода являются основным инструментом для интерактивной разработки, анализа данных, машинного обучения и других вычислительных задач в Jupyter Notebook.

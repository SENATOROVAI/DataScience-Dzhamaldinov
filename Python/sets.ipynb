```python
# множество это совокупность объектов, объекты у множества уникальны 
# структура данных множество
# неявное создание объекта класса set с помощью фигурных скобок 
my_fruits = {"apple", "orange", "apple"}
# явное создание объекта 
my_fruits_obj = set(["apple", "orange", "apple"])
your_fruits = {"orange", "banana", "pear"} 
# print(my_fruits)
number_set = {1, 2, 3, 1, 1, 1}
# print(number_set)
# показывает имя класса или тип данных
# type(my_fruits)

# union_fruits = my_fruits.union(your_fruits)
union_fruits = my_fruits | your_fruits
union_fruits
```

```python
# Для объединения множеств можно использовать оператор `|`,
# оба операнда должны быть объектами типа set
our_fruits = my_fruits | your_fruits
print(our_fruits)

```

```python
# Также можно использовать ментод union.
# Отличие состоит в том, что метод union принимает не только
# объект типа set, а любой iterable-объект
you_fruit_list: list = list(your_fruits)
our_fruits_second: set = my_fruits.union(you_fruit_list)
print(our_fruits_second)

```
